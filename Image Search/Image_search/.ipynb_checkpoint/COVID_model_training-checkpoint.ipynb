## 2. Data pre-processing

Before the model can be created with Resnet-50, the data must be split into classifications and then further into training, testing and validation sets.

Following the READ.ME document, your data set should already be split into classifications and stored in the "data" folder, and so the following cell will further split each classification into the 3 sets:

```python
# split data in a new folder named data-split
splitfolders.ratio("COVID-train-data", output="COVID-train-split", seed=1337, ratio=(0.7, 0.2, 0.1), group_prefix=None, move=False)
```

## 3. Creating Keras Data Generators

In the following section, you must input the labels for each classification for the model to be trained with.

Simply input these into the following cell before continuing:

```python
# define classes name
class_names = ['COVID','Lung_Opacity','Normal','Viral_Pneumonia']  # replace each class_name with all class names required
```

```python
datagen = ImageDataGenerator()
```

```python
# training data
train_generator = datagen.flow_from_directory( 
    directory="COVID-train-split/train/", 
    classes = class_names,
    target_size=(299, 299),  
    batch_size=32, 
    class_mode="binary", 
)
```

```python
# validation data
valid_generator = datagen.flow_from_directory( 
    directory="COVID-train-split/val/", 
    classes = class_names,
    target_size=(299, 299), 
    batch_size=32, 
    class_mode="binary", 
)
```

```python
# test data
test_generator = datagen.flow_from_directory( 
    directory="COVID-train-split/test/", 
    classes = class_names,
    target_size=(299, 299), 
    batch_size=32, 
    class_mode="binary", 
)
```

## 4. Building the model

```python
# ResNet50 model
res_net = ResNet50(include_top=False, weights='imagenet', input_shape=(299,299,3))
for layer in res_net.layers:
    layer.trainable = False
```

```python
# build the entire model
x = res_net.output
x = layers.GlobalAveragePooling2D()(x) 
x = layers.Dropout(0.5)(x) 
x = layers.Dropout(0.5)(x) 
x = layers.Dropout(0.5)(x) 
x = layers.Dropout(0.5)(x)
predictions = layers.Dense(2048, activation='softmax')(x)
model = Model(inputs = res_net.input, outputs = predictions)
```

## 5. Training the model

```python
# define training function
def trainModel(model, epochs, optimizer):
    batch_size = 32
    model.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model.fit(train_generator, validation_data=valid_generator, epochs=epochs, batch_size=batch_size)
```

```python
# launch the training
model_history = trainModel(model = model, epochs = 10, optimizer = "Adam")
```

The folllowing cells will create graphs allowing you to visualise the loss and accuracy for the model.

Loss curves:

```python
loss_train_curve = model_history.history["loss"]
loss_val_curve = model_history.history["val_loss"]
plt.plot(loss_train_curve, label = "Train")
plt.plot(loss_val_curve, label = "Validation")
plt.legend(loc = 'upper right')
plt.title("Loss")
plt.show()
```

Accuracy Curves:

```python
acc_train_curve = model_history.history["accuracy"]
acc_val_curve = model_history.history["val_accuracy"]
plt.plot(acc_train_curve, label = "Train")
plt.plot(acc_val_curve, label = "Validation")
plt.legend(loc = 'lower right')
plt.title("Accuracy")
plt.show()
```

## 6. Evaluating the model

The following cell will evaluate the loss and accuracy of the created model:

```python
test_loss, test_acc = model.evaluate(test_generator)
print("The test loss is: ", test_loss)
print("The best accuracy is: ", test_acc*100)
```

### IMPORTANT

The following cell will search your data folder for files that are incompatible with Tensorflow. These files will then be deleted from the data folder, so it is important that if you want to keep all of these images that you have the data set saved elsewhere as a backup.

```python
from pathlib import Path
import imghdr

data_dir = "data/"
image_extensions = [".png", ".jpg", ".jpeg"]  # add there all your images file extensions

img_type_accepted_by_tf = ["bmp", "gif", "jpeg", "png"]
for filepath in Path(data_dir).rglob("*"):
    if filepath.suffix.lower() in image_extensions:
        img_type = imghdr.what(filepath)
        if img_type is None:
            print(f"{filepath} is not an image")
            print(f"Deleting {filepath} from data folder")
            os.remove("{filepath}")
        elif img_type not in img_type_accepted_by_tf:
            print(f"{filepath} is a {img_type}, not accepted by TensorFlow")
            print(f"Deleting {filepath} from data folder")
            os.remove("{filepath}")
```

## 9. Loading image data

Defining list of paths to extracted images:

```python
def extract_file_paths_from_folder(parent_dir: str) -> dict:
    image_paths = {}
    for sub_folder in os.listdir(parent_dir):
        sub_dir = os.path.join(parent_dir, sub_folder)
        image_paths[sub_folder] = [
            os.path.join(sub_dir, file) for file in os.listdir(sub_dir)
        ]
    return image_paths
```

## 11. Storing Embeddings in KDB.AI

### Defining KDB.AI session

As part of the README document, it will be stated that you must have KDB.AI setup before starting use of this document. Should this be the case, the following cell will define a KDB.AI session:

```python
import kdbai_client as kdbai
session = kdbai.Session(endpoint='http://localhost:8082')
```

### Defining vector DB table schema:

```python
image_schema = {
    "columns": [
        {"name": "source", "pytype": "str"},
        {"name": "class", "pytype": "str"},
        {
            "name": "embedding",
            "vectorIndex": {"dims": 2048, "metric": "L2", "type": "hnsw"},
        },
    ]
}
```

### Creating vector DB table:

```python
# ensure the table does not already exist
try:
    session.table("yourTable").drop()
    time.sleep(5)
except kdbai.KDBAIException:
    pass
```

```python
table = session.create_table("yourTable", image_schema)
```

### Adding embedded data to the table

This next stage requires some added steps depending on how large your embedding vector data set is. The "insert" command that will be used in this stage can only insert a certain number of bytes, with a general rule of thumb that 10mb is the maximum amount of data that can be inserted at once.

The following cell will provide a rough estimate of how many megabytes your embedding vector data set is made up of:

```python
# convert bytes to MB
embedded_df.memory_usage(deep=True).sum() / (1024**2)
```

If the data set is comfortably below the 10mb limit, then you should be able to insert the embeddings into the table in one step using the following:

```python
table.insert(embedded_df)
```

Should the data set be larger than 10mb, you will need to divide the data set into smaller parts. This can be done using the following cells.

First of all, it is important to get a rough estimate of how many items will be in each block. This can be done with the following cell:

```python
import math
megab = embedded_df.memory_usage(deep=True).sum() / (1024**2)
megabs = megab/10
blocks = len(embedded_df)/megabs
math.floor(blocks)
```

The previous cell will have provided a rough estimate for an upper limit to the amount of items within each block. The following cell will break the data set into blocks of a specified size. Try this with the estimated block size provided.

```python
# Yield successive n-sized 
# chunks from l. 
def divide_chunks(l, n): 
      
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n] 
  
# How many elements each 
# list should have 
n = 500
```

```python
embedded_df_split = list(divide_chunks(embedded_df, n))
```

Now that the data set has been split into smaller blocks, it can be inserted into the KDB.AI table using the following cell:

```python
for i in range(len(embedded_df_split)):
    table.insert(embedded_df_split[i])
```

Should you still be returned with an error, try breaking the table into smaller blocks than you are currently using and eventually the blocks will be small enough to be inserted into the table.

You can now verify that the data has been inserted into the table with the following cell:

```python
table.query()
```

You can make specific queries into the table using the following:

```python
table.query(filter=[("like", "class", "*RedRot*")])
```

### Plotting the most similar images:

```python
def plot_test_result_with_8NN(test_file: str, neighbors: pd.Series) -> None:
    # create figure
    _, ax = plt.subplots(nrows=3, ncols=3, figsize=(12, 7))
    axes = ax.reshape(-1)

    # plot query image
    plot_image(axes[0], test_file, "Test")

    # plot nearest neighbors
    for i, (_, value) in enumerate(neighbors.items(), start=1):
        plot_image(axes[i], value, f"{i}-NN")
```
